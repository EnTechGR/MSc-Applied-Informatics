# Γ: Στατιστική Συμπερασματολογία-Έλεγχοι Υποθέσεων (σε επίπεδο σημαντικότητας α=5%)


````python
import scipy.stats as stats
````



# Γ.1
 Να διατυπώσετε τον κατάλληλο έλεγχο υποθέσεων για τον ισχυρισμό ότι ο μέσος
χρόνος ομιλίας ΤΙΜΕ στον αντίστοιχο πληθυσμό διαφέρει στατιστικά σημαντικά από
τις 100 μονάδες. Σχολιάστε.

H_0: μ=100
vs
H_1: μ#100



````python
stats.ttest_1samp(TIME,alternative='two-sided', popmean=100)
````

Eπειδή p-value~0<alpha, alpha=0.05, απορρίπτουμε την Η_0 σε επίπεδο σημαντικότητας 5%.

Η_0:μ<=100 vs H_1:μ>100

````python
stats.ttest_1samp(TIME,alternative='greater', popmean=100)
````

Επιλέγουμε την εναλλακτική υπόθεση alternative='greater', που υποδεικνύει ότι μας ενδιαφέρει να δούμε εάν η μέση τιμή του δείγματος είναι μεγαλύτερη από την προσδοκώμενη τιμή popmean=100.
Eπειδή p-value~0<alpha, alpha=0.05, απορρίπτουμε την Η_0 σε επίπεδο σημαντικότητας 5%.

Ελεγχος ισοτητας διασπορών του πληθυσμού των ανδρών σε σχεση με των γυναικών

Η_0:σ_1^2=σ_2^2 vs H_1:σ_1^2#σ_2^2

s ή σ είναι η διασπορά

````python
# converting the list to array
x = np.array(TIME1)
y = np.array(TIME2)

# calculate variance of each group
print(np.var(TIME1), np.var(TIME2))

def f_test(TIME1, TIME2):
    f = np.var(TIME1, ddof=1)/np.var(TIME2, ddof=1)
    nun = x.size-1
    dun = y.size-1
    p_value = 1-stats.f.cdf(f, nun, dun)
    return f, p_value

# perform F-test
f_test(x, y)
````

Eπειδή p-value=0.5881211355984333>>>alpha, alpha=0.05, δεν απορρίπτουμε την Η_0 σε επίπεδο σημαντικότητας 5%. Συνεπώς, δεν μπορούμε να απορρίψουμε τη μηδενική υπόθεση και μπορούμε να συμπεράνουμε ότι οι διασπορές του χρόνου για τους άνδρες και τις γυναίκες δεν είναι στατιστικά διαφορετικές.
